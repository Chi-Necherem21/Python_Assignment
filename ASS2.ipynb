```python
# Civil Servant Retirement Calculator

# collect user Inputs 
title = input("Enter Title (Mr/Mrs/Miss/Dr): ")
first_name = input("Enter First Name: ")
last_name = input("Enter Last Name: ")

commenced_year = int(input("Enter Year Commenced Civil Service: "))
birth_year = int(input("Enter Birth Year: "))
current_year = int(input("Enter Current Year: "))

#  retirement age
RETIREMENT_AGE = 60

# Calculations of current age
current_age = current_year - birth_year
years_of_service = current_year - commenced_year

retirement_year = birth_year + RETIREMENT_AGE
years_left = retirement_year - current_year

# results
print("      CIVIL SERVICE RETIREMENT REPORT")
print("==============================================")

print(f"Hello {title} {first_name} {last_name}.")

print(f"You are {current_age} years old.")
print(f"You will be retiring in the year {retirement_year} at the age of {RETIREMENT_AGE}.")
print(f"You have spent {years_of_service} years of active service.")

if years_left > 0:
    print(f"You have {years_left} years of active service left.")
elif years_left == 0:
    print("Congratulations! You are due for retirement this year.")
else:
    print(f"You retired {-years_left} year(s) ago.")

print("Thank you for your service.")
```
