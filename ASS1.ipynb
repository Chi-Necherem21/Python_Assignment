```python
# BMI Calculator

# Receive inputs
weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in metres: "))

# Calculate BMI
bmi = round(weight / (height ** 2), 2)

# Categorise BMI using IF statements
if bmi < 18:
    print(f"Your body mass index is {bmi}. You are Underweight!")
elif 18 <= bmi <= 25:
    print(f"Your body mass index is {bmi}. You are Normal Weight!")
else:
    print(f"Your body mass index is {bmi}. You are Overweight or Obese!")

# Display BMI Chart
msg = """
Body Mass Index Chart:
Between 18 - 25 = Normal Weight
Less than (<) 18 = Underweight
Greater than (>) 25 = Overweight OR Obese
"""

print(msg)
```
